**Predict sentiment using Textblob. How many positive titles (original_title) are there [books.csv] ? (cut-off >=0) N**

```python
book1=book.dropna(subset=['original_title'])
```

```python
book1.shape
```

```python
from textblob import TextBlob
# Define a function to calculate sentiment
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Apply the sentiment function to the original_title column
book1['sentiment'] = book1['original_title'].apply(get_sentiment)

# Filter for positive sentiments (cut-off >= 0)
positive_titles = book1[book1['sentiment'] >= 0]

# Count the number of positive titles
num_positive_titles = positive_titles.shape[0]

print(f"Number of positive titles: {num_positive_titles}")
```
